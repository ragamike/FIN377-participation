```python
diamonds = sns.load_dataset('diamonds')
diamonds # notice shape, unit, key, etc...
```

```python
# my turn: lets do the usual immediate explorations including the categorical vars
print(diamonds.head())
print(diamonds.tail())
print(diamonds.describe())
```

```python
# my turn: explore carats - how many are 0.99 vs 1 carat? why?
print(len(diamonds.query('carat == 0.99')))
print(len(diamonds.query('carat == 1.0')))

#nobody wants a diamonds thats just short of being 1.0, they'd rather get the whole carat
```

```python
#explore dist of x, y, z: what did you learn? which is width, length, depth?
print(diamonds['x'].describe())
print()
print(diamonds['y'].describe())
print()
print(diamonds['z'].describe())
```

Length and width (presumably x and y) are close to being the same size as one another, while depth (z) is usually much smaller

```python
# your turn: explore price - is there anything unusual? (HINT: try many bin widths)
diamonds['price']
sns.distplot(diamonds['price'], bins = 100)
#the price of diamonds is heavily concentrated around a specific price point
```

```python
# my turn: how is carat related to price?
diamonds.corr()
#carat and price are highly correlated at 0.921
```

```python
# your turn: how is price related to cut?
diamonds.groupby(['cut']).corr()
#cut becomes more important as the carat of the diamond increases
```
