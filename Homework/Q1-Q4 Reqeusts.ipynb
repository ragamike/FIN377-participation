# Q3: What was the number of Yen per Euro on Jan 2, 2018?

```python
new_url = 'https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2018-01-02'
r = requests.get(new_url)
pd.DataFrame(r.json())
r.json()['rates']
r.json()['rates']['2018-01-02']['JPY']
```

# Q4: Bonus, prof can show: Get a time series of EURJPY from 2018 through 2019.

```python
new_url = 'https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2019-12-31'
r = requests.get(new_url)

r.json()
pd.DataFrame(r.json())
df = pd.DataFrame(r.json())
df['EURJPY'] = df['rates'].apply(lambda x: x.get('JPY'))
df
```
