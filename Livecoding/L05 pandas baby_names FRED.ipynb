## Notes
 
split, apply, combine

1. take data set and split it into 2 (ex: john's split by year)
2. apply the function (ex: `.sum()`)
3. combine the split data sets

```python
# count john's over time

somenames = baby_names.loc[baby_names['name'] == "John"]
combined = somenames.groupby(['name', 'year'])['count'].sum().reset_index()     # groupby variables at the name time

# remember to reset index before doing this! turns year into a variable
sns.lineplot(data = combined, x = 'year', y = 'count')
```

```python
# arya and sansa

# | = or

names_i_want = (baby_names['name'] == "Sansa") | (baby_names['name'] == "Arya")
names_i_want.sum()

somenames = baby_names.loc[names_i_want]
combined = somenames.groupby(['name', 'year'])['count'].sum().reset_index()
sns.lineplot(data = combined, x = 'year', y = 'count', hue = 'name')
```

```python
# query!!!
# needs a string with T/F conditions

somenames = baby_names.query('name in ["Arya", "Sansa"]')
combined = somenames.groupby(['name', 'year'])['count'].sum().reset_index()
sns.lineplot(data = combined, x = 'year', y = 'count', hue = 'name')
```

```python
# i don't want "somenames"

(     # () means that everything in the middle is one line of code 
baby_names
    .query('name in ["Arya", "Sansa"]')     # get a subsample of the dataset
    .groupby(['name', 'year'])['count'].sum().reset_index()     # combine M and F
    
    .pipe(     # pipe in a function to apply to the object
               # syntax: <function>, arg1, arg2, ...
        (sns.lineplot, 'data'), x = 'year', y = 'count', hue = 'name'
    
    # call the function and tell it that the object in memory goes to the parameter 'data'
    
    )
)
```
